```python
# Module for plotting ARIMA results only!
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Mon Nov 27 2017

@author: jshermeyer
"""

import pandas as pd
import geopandas as gpd
import shapely.wkt
import seaborn as sns
from ARIMA import run_plot_TS
sns.set(color_codes=True)
%matplotlib inline


# Run plotting from CSV output ONLY

input_CSV = "/Users/jshermeyer/CosmiqGit/CometTS_dev/VIIRS_Sample/San_Juan_ARIMA_Output.csv"


def gen_plots(input_csv):
    df = pd.read_csv(input_csv)
    df = df.sort_values(['date'])
    geometry = df['geometry'].map(shapely.wkt.loads)
    crs = {'init': 'epsg:2263'}
    gdf = gpd.GeoDataFrame(df, crs=crs, geometry=geometry)
    gdf['date'] = pd.to_datetime(gdf['date'], infer_datetime_format=True)
    # run_plot(gdf, y_label="NDVI", title_label= "NDVI over time - ID: ")
    run_plot_TS(gdf)


gen_plots(input_CSV)
```
